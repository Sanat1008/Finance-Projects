```python
import pandas as pd 
import matplotlib.pyplot as plt
data = pd.read_csv("nifty (1).csv")
data.head(5)
plt.plot(data["Date"],data["High"],color="red",label="Line Graph")
plt.scatter(data["Date"],data["High"],color="blue",label="Scatter")
#plt.bar(data["High"],data["Low"],color="blue,label="bar)
#plt.barh(data["High"],data["Low"],color=["green","blue"])
plt.legend()
plt.xlabel("Date",color="green")
plt.ylabel("High",color="red")
plt.title("High",color="red")
plt.title("High vs Low",color="green")
plt.show()
```
